```python
#Importing libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading file

df = pd.read_csv("C:/Users/hherre13/OneDrive - JNJ/Documents/Hector/Python Total para data science/Día 9/Cuadernos para Prácticas/Datos+Meteorológicos_Arg_2023.csv")
df
```

```python
df["Fecha"] = pd.to_datetime(df["Fecha"], format='%d/%m/%Y')
```

```python
#Crear dos variables: Ciudades y meses
lista_ciudades = []

for c in df["Ciudad"]:
    if c not in lista_ciudades:
        lista_ciudades.append(c)

lista_ciudades
```

```python
#Diccionario con todos los numeros por mes

dict_meses = {1: "Enero",
             2:"Febrero",
             3:"Marzo",
             4:"Abril",
             5:"Mayo",
             6:"Junio",
             7:"Julio",
             8:"Agosto",
             9:"Septiembre",
              10:"Octubre",
              11:"Noviembre",
              12:"Diciembre"}
```

```python
def consultar_temperaturas():
    while True:
        #Solicitar ciudad
        print("\nCiudades disponibles: ", lista_ciudades)
        ciudad_elegida = input("Elija la ciudad de la lista: ").title()
        
        #Solicitar el mes
        mes_elegido = int(input("Elija un numero de mes (por ejemplo Enero =1 )"))
        
        #Validar los inputs
        if ciudad_elegida not in lista_ciudades or mes_elegido not in range(1,13):
            print("Ciudad o mes no validos. Intentalo nuevamente")
            continue
        
        #"crear el dataframe filtrado"
        datos_ciudad_mes = df[(df["Ciudad"] == ciudad_elegida) &(df["Fecha"].dt.month ==  mes_elegido)]
        
        #"Graficamos las temperaturas del df"
        plt.figure(figsize=(10,6))
        plt.plot(datos_ciudad_mes["Fecha"],datos_ciudad_mes["Temperatura Maxima"], label="Maxima",color="red")
        plt.plot(datos_ciudad_mes["Fecha"],datos_ciudad_mes["Temperatura Minima"], label="Minima",color="blue")
        plt.title(f"Temperaturas en {ciudad_elegida} durante el mes de {dict_meses[mes_elegido]}")
        plt.xlabel("Fecha")
        plt.ylabel("Temperaturas")
        plt.legend()
        plt.xticks(rotation=45)
        plt.show()
            
        #Preguntar al usuario si quieren continuar
        otra_consulta = input("Desas continuar consultando")
        if otra_consulta.lower() != "s":
            break
```

```python
consultar_temperaturas()
```
