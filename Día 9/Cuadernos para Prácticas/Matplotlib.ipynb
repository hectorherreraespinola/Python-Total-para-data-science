```python
#Ejercicio 1 
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({"x":range(1,11),
                "y": [i **2 for i in range(1,11)]})


fig, ax = plt.subplots()

ax.stem(df["x"],df["y"]);


ax.set(xlim=(0, 11), xticks=range(1, 11),
       ylim=(0, 110), yticks=range(1, 10))
plt.title( "Relacin entre Nmeros y sus Cuadrados")
plt.xlabel("Nmero (x)")
plt.ylabel("Cuadrado del nmero (x^2)");
```

```python
#Ejercicio 2


x = [1, 2, 3, 4, 5]
y1 = [6, 1, 4, 4, 8]
y2 = [4, 6, 3, 5, 8]
y3 = [7, 9, 9, 2, 7]

# Crear el grfico apilado
fig, ax = plt.subplots()
ax.stackplot(x, y1, y2, y3, labels=['Serie y1', 'Serie y2', 'Serie y3'], colors=['red', 'blue', 'green'])

# Aadir ttulo y etiquetas de los ejes
ax.set_title('Distribucin acumulativa de valores por serie de tiempo')
ax.set_xlabel('Tiempo')
ax.set_ylabel('Valores Acumulados')

# Aadir leyenda
ax.legend(loc='upper left')

# Mostrar el grfico
plt.show()
```

```python
#Ejercicio 3

import numpy as np
import matplotlib.pyplot as plt

# Establecer la semilla para la generacin de datos
np.random.seed(0)

# Generar los datos para cada muestra usando numpy.random.normal
data1 = np.random.normal(loc=0, scale=0.5, size=100)
data2 = np.random.normal(loc=2, scale=1, size=100)
data3 = np.random.normal(loc=-2, scale=1.5, size=100)

# Combinar los datos en una lista
data = [data1, data2, data3]

# Crear el grfico de violn
fig, ax = plt.subplots()
ax.violinplot(data)

# Establecer las etiquetas en el eje x para identificar cada una de las muestras
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(['Muestra 1', 'Muestra 2', 'Muestra 3'])

# Aadir un ttulo al grfico
ax.set_title("Comparacin de distribuciones en Violin Plot")

# Aadir etiquetas a los ejes
ax.set_xlabel('Muestras')
ax.set_ylabel('Valores')

# Mostrar el grfico
plt.show()
```
