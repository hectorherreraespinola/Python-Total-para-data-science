Merge

```python
df1 = pd.DataFrame({"ID":[1,2,3],
                   "Nombre":["Ana","Luis","Carlos"]})

df2 = pd.DataFrame({"ID":[1,2,4],
                   "Edad":[25,30,22]})
```

```python
df1
```

```python
df2
```

```python
df_combinado = pd.merge(df1,df2,on= "ID", how="inner")
df_combinado
```

```python
df_combinado = pd.merge(df1,df2,on= "ID", how="outer")
df_combinado
```

```python
df_combinado = pd.merge(df1,df2,on= "ID", how="left")
df_combinado
```

```python
df_combinado = pd.merge(df1,df2,on= "ID", how="right")
df_combinado
```

```python
df_indexado = pd.merge(df1,df2, left_index= True, right_index= True)
df_indexado
```

```python
# CreaciÃ³n del DataFrame df_a
df_a = pd.DataFrame({
    'id': [1, 2, 3],
    'Nombre': ['Ana', 'Beto', 'Carla']
})
df_a.set_index('id', inplace=True)

# CreaciÃ³n del DataFrame df_b
df_b = pd.DataFrame({
    'id': [1, 2, 3],
    'Edad': [25, 30, 35]
})
df_b.set_index('id', inplace=True)

df_combinado = df_a.join(df_b)
```

```python
productos_df = pd.DataFrame({
    'ProductoID': [1, 2, 3, 4],
    'Nombre': ['Producto 1', 'Producto 2', 'Producto 3', 'Producto 4'],
    'Precio': [100, 150, 200, 250]
}).set_index('ProductoID')

categorias_df = pd.DataFrame({
    'CategoriaID': [1, 2],
    'NombreCategoria': ['CategorÃ­a 1', 'CategorÃ­a 2']
}).set_index('CategoriaID')


df_combinado = productos_df.join(categorias_df,how="left")

df_combinado
```

```python
#Ejercicio 3

productos_df = pd.DataFrame({
    'ProductoID': [1, 2, 3, 4],
    'Nombre': ['Producto 1', 'Producto 2', 'Producto 3', 'Producto 4'],
    'Precio': [100, 150, 200, 250]
})

productos_df = productos_df.set_index('ProductoID')

categorias_df = pd.DataFrame({
    'CategoriaID': [1, 2, 3],
    'NombreCategoria': ['CategorÃ­a 1', 'CategorÃ­a 2', 'CategorÃ­a 3']
}).set_index('CategoriaID')



df_combinado = productos_df.join(categorias_df,how="right")

df_combinado
```

Concatenar Dataframes: Permite unir dataframes

```python
df1 = pd.DataFrame({"Nombre":["Juan","Gabriela","Elena"], "Edad":[23,31,21]})
df1
```

```python
df2 = pd.DataFrame({"Nombre":["Carmela","Max","Laura"], "Edad":[34,25,29]})
df2
```

```python
df_concatenado = pd.concat([df1,df2]) ##Eje de las columnas
df_concatenado
```

```python
df_concatenado = pd.concat([df1,df2],axis =1 ) 
df_concatenado
```

```python
df_concatenado = pd.concat([df1,df2], ignore_index=True)
df_concatenado
```

```python
df_concatenado = pd.concat([df1,df2], keys=["df1","df2"]) 
df_concatenado
```
