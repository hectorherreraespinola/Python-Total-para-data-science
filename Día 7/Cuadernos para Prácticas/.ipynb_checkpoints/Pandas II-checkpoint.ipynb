Merge

```python
df1 = pd.DataFrame({"ID":[1,2,3],
                   "Nombre":["Ana","Luis","Carlos"]})

df2 = pd.DataFrame({"ID":[1,2,4],
                   "Edad":[25,30,22]})
```

```python
df1
```

```python
df2
```

```python
df_combinado = pd.merge(df1,df2,on= "ID", how="inner")
df_combinado
```

```python
df_combinado = pd.merge(df1,df2,on= "ID", how="outer")
df_combinado
```

```python
df_combinado = pd.merge(df1,df2,on= "ID", how="left")
df_combinado
```

```python
df_combinado = pd.merge(df1,df2,on= "ID", how="right")
df_combinado
```

```python
df_indexado = pd.merge(df1,df2, left_index= True, right_index= True)
df_indexado
```

```python
#Ejercicio 1

libros = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'titulo': ['El Quijote', 'La Odisea', 'Cien AÃ±os de Soledad', 'El Principito']
})

autores = pd.DataFrame({
    'ID': [1, 2, 3, 5],
    'nombre': ['Miguel de Cervantes', 'Homero', 'Gabriel GarcÃ­a MÃ¡rquez', 'J.K. Rowling']
})


df_libros = pd.DataFrame(libros)
df_autores = pd.DataFrame(autores)


libros_autores = pd.merge(df_libros,df_autores,on="ID", how="outer")

libros_autores
```

```python
#Ejercicio#2
cursos = pd.DataFrame({
    'curso_id': [101, 102, 103],
    'nombre_curso': ['IntroducciÃ³n a Python', 'Data Science con Python', 'Machine Learning Avanzado']
})

inscripciones = pd.DataFrame({
    'curso_id': [102, 102, 101, 104],
    'fecha_inscripcion': ['2023-01-15', '2023-02-01', '2023-01-20', '2023-03-05']
})

df_cursos = pd.DataFrame(cursos)
df_inscripciones = pd.DataFrame(inscripciones)


cursos_inscripciones = pd.merge(df_inscripciones,df_cursos,on="curso_id",how="left")

cursos_inscripciones
```

```python
productos = pd.DataFrame({
    'ID': [10, 11, 12],
    'Nombre': ['Teclado', 'Mouse', 'Monitor'],
    'Marca': ['Logitech', 'Razer', 'Dell']
})

reviews = pd.DataFrame({
    'ID': [10, 11, 12],
    'CalificaciÃ³n': [5, 4, 4],
    'Comentario': ['Excelente producto', 'Buen producto', 'Satisfecho']
})

df_productos = pd.DataFrame(productos)
df_reviews = pd.DataFrame(reviews)

productos_review = pd.merge(df_productos,df_reviews, left_index = True, right_index = True)

productos_review
```
