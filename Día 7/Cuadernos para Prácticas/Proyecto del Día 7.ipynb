```python
df_concatenado["Fecha"] = pd.to_datetime(df_concatenado["Fecha"])
```

```python
#Ejercicio 4
#Producto mas vendido

producto_mayor_ventas= df_concatenado.groupby("Producto")["Cantidad"].sum()
producto_mayor_ventas = producto_mayor_ventas.sort_values(ascending=False)
producto_mayor_ventas.head(1)
```

```python
meses = []
for i in df_concatenado["Fecha"]:
    meses.append(i.month)
df_concatenado["Meses"] = meses
df_concatenado
```

```python
ventas_por_mes = df_concatenado.groupby("Meses")["Total Venta"].sum().sort_values(ascending=False)
ventas_por_mes.head(1)
```
