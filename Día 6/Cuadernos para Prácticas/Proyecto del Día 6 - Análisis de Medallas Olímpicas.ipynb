```python
#Análisis de Medallas de Oro por País: Realiza las operaciones que sean necesarias para identificar cuáles 
#fueron los 3 países con más medallas de Oro en total 
#(vas a necesitar investigar los métodos de dataframes para encontrar cuál te permite ordenar 
#los valores de mayor a menor o viceversa).

df_medallas = df_medallas.groupby("Pais").sum()
df_medallas = df_medallas.sort_values("Oro", ascending= False)
df_medallas.head(3)

```
